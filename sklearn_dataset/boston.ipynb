```python
import sklearn.datasets as ds;
import sklearn.linear_model as lm;
import matplotlib.pyplot as plot;
boston=ds.load_boston()
data=boston.data;
target=boston.target;
#这里获取的是一个二维数组,并且过滤掉target<50的，其实就是找出target<50的索引然后取对应位置的数据
rm=data[:,5:6][target<50]
target=target[target<50]
print(rm)
plot.scatter(rm[:,-1],target,c="red")
plot.show();
# 注意线性回归，二维数组有几列就是有几个特征项
lr=lm.LinearRegression();
lr.fit(rm,target);
print(lr.coef_)
#截距
print(lr.intercept_)
```
